Q3. Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
# i. x = np.array([1,2,3,4,5,1,2,1,1,1])
x = np.array([1,2,3,4,5,1,2,1,1,1])

value = np.bincount(x).argmax()
indices = np.where(x == value)[0]

print("i. x = np.array([1,2,3,4,5,1,2,1,1,1])")
print("Most frequent value:", value)
print("Indices:", indices)


#ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])
print("\nii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])")

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])
counts = np.bincount(y)
max_freq = counts.max()
values = np.where(counts == max_freq)[0]

for v in values:
    print("Value:", v)
    print("Indices:", np.where(y == v)[0])


```

Q4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
find

i. Sum of all elements

ii. Sum of all element’s row-wise

iii. Sum of all element’s column-wise

```python
#Q4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), find i. Sum of all element
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

#i. Sum of all elements
sum_all= np.sum(gfg)
print("Sum of all elements:", sum_all)

#ii. Sum of all element’s row-wise
sum_row= np.sum(gfg, axis=1)
print("Sum of all elements row-wise:\n", sum_row)

#iii. Sum of all element’s column-wise
sum_col= np.sum(gfg, axis=0)
print("Sum of all elements column-wise:", sum_col)
```
